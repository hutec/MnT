```python
%matplotlib inline

import matplotlib.pylab as plt
import pandas as pd
import matplotlib.cm as cm
import matplotlib.colors as colors
from matplotlib.font_manager import FontProperties

import numpy as np
```

```python
colorlist = list(colors.ColorConverter.colors.keys())
```

```python
cmap = plt.get_cmap('gnuplot')
colors_arr = [cmap(i) for i in np.linspace(0, 1, 16)]
```

```python
color_arr = cm.rainbow(np.linspace(0, 1, 17))  
```

```python
df = pd.read_excel("MnT_U4.xlsx")
```

```python
colormap = cm.viridis
colorlist = [colors.rgb2hex(colormap(i)) for i in np.linspace(0, 0.9, len(df))]
plt.figure(figsize=(20,20))
i = 0
for k, v in df.iterrows():
    plt.scatter(v['Ressourcenstaerke'], v['Attraktivitaet'], label=v.name, color=color_arr[i], s=400)
    if v.name == "Algorithmen":
                plt.annotate(v.name, (v['Ressourcenstaerke'], v['Attraktivitaet']), xytext=(4, 2.2), fontsize=20, 
                             arrowprops=dict(facecolor='black', shrink=0.05))
    else:
         plt.annotate(v.name, (v['Ressourcenstaerke'], v['Attraktivitaet']), fontsize=20)
    i += 1
plt.xlabel("Ressourcenstaerke", fontsize=20)
plt.xlim(-0.25,4.25)
plt.ylim(-0.25,4.25)
plt.ylabel("Attraktivitaet", fontsize=20)
plt.legend(loc='lower center', bbox_to_anchor=(0.5, 1.05),
          ncol=3, fancybox=True, shadow=True)
plt.plot([0, 3],[3, 0], color="brown")
plt.plot([1, 4],[4, 1], color="brown")
plt.savefig("portfolio.png", bbox_inches='tight')
```
